```python
# ixプロパティの勉強

import pandas_datareader.data as pdr

start = "1949/5/6"
end = "2016/9/30"

# pandas_datareaderを使って、yahooからNIKKEI225のデータを読み込み <DataFrame>
price = pdr.DataReader("^N225", 'yahoo', start, end)
print('\n--- 1. N225 from yahoo <DataFrame> ---')
print(price.head(5))

# yahooから取得したNIKKEI225のデータから 2016.01.01以降を取得 <DataFrame>
print('\n--- 2. slice N225 data at 20160101 and get after 20160101  <DataFrame> ---')
price_from20160101 = price.ix["2016/1/1":]
print(price_from20160101.head(5))

# yahooから取得したNIKKEI225のデータから 2016.01.01以降 高値(High)のみを表示 <Series>
print('\n--- 3. price_from20160101.High <Series> ---')
print(price_from20160101.High.head(5))

# yahooから取得したNIKKEI225の2016.01.01以降のデータから 2016年のレコードのindex1を取得する。つまりHigh <Series>
print('\n--- 4. price_from20160101.ix["2016",1] <Series> ---')
print(price_from20160101.ix["2016",1].head(5))

# yahooから取得したNIKKEI225の2016.01.01以降のデータから 2016年のレコードの"High"の カラムを取得 <Series>
print('\n--- 5. price_from20160101.ix["2016","High"] <Series> ---')
print(price_from20160101.ix["2016","High"].head(5))

# yahooから取得したNIKKEI225の2016.01.01以降のデータから 2016年のレコードの"High"と"Row"の カラムを取得 <DataFrame>
print('\n--- 6. price_from20160101.ix["2016",["High","Low"]] <DataFrame> ---')
print(price_from20160101.ix["2016",["High","Low"]].head(5))

# yahooから取得したNIKKEI225の2016.01.01以降のデータから 行列番号を指定して値を取得 <Value>
print('\n--- 7. price_from20160101.ix[0,1] <Value> ---')
print(price_from20160101.ix[0,1])

# yahooから取得したNIKKEI225の2016.01.01以降のデータから 行番号をリストで指定してシリーズを取得 <Series>
print('\n--- 8. price_from20160101.ix[[0],1] <Series> ---')
print(price_from20160101.ix[[0],1])

# yahooから取得したNIKKEI225の2016.01.01以降のデータから 行列番号をそれぞれリストで指定してデータフレームを取得 <DataFrame>
print('\n--- 9. price_from20160101.ix[[0],[1] <DataFrame> ---')
print(price_from20160101.ix[[0],[1]])

# yahooから取得したNIKKEI225の2016.01.01以降のデータから 行の値と列番号を指定して値を取得 <Value>
print('\n--- 10. price_from20160101.ix["2016/1/4",1] <Value> ---')
print(price_from20160101.ix["2016/1/4",1])

# yahooから取得したNIKKEI225の2016.01.01以降のデータから 行の値を指定、列番号をリスト指定してシリーズを取得 <Series>
print('\n--- 10. price_from20160101.ix["2016/1/4",[1,2]] <Series> ---')
print(price_from20160101.ix["2016/1/4",[1,2]])
```
