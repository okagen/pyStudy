```python
# NetworkX : ネットワーク構造を分析するライブラリ
# グラフ理論のアルゴリズムを提供
%matplotlib notebook
import networkx as nx
import matplotlib.pyplot as pp

# グラフオブジェクト作成
G = nx.Graph()

# ノードを追加
#G.add_node(1) #1つ追加
G.add_nodes_from(['A','B','C','D']) # リストから追加
# edgeを追加
G.add_weighted_edges_from([('A', 'B',5),('A', 'C', 3),('B', 'C',6),('B', 'D',8),('C', 'D',9)])

# matplotlibで描画
nx.draw(G, with_labels=True)
pp.show()

# ダイクストラ法 dijkstra_path 採点経路を求めるアルゴリズム
p1 = nx.dijkstra_path(G,'A','D')
print(p1)

# ダイクストラ法 dijkstra_path_length 採点経路の所要時間
p2 = nx.dijkstra_path_length(G,'A','D')
print(p2)

# 最小全域木
mst = nx.minimum_spanning_tree(G)
print(mst.edges())
```
