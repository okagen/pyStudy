```python
# resumpleメソッドの勉強
import pandas_datareader.data as pdr

start = "1949/5/6"
end = "2016/9/30"

# very useful sight
# http://www.webpages.uidaho.edu/~stevel/504/Pandas%20DataFrame%20Notes.pdf

# pandas_datareaderを使って、yahooからNIKKEI225のデータを読み込み
n225 = pdr.DataReader("^N225", 'yahoo', start, end)

# 2016年の高値と低値の値を表示
print('\n--- 1. n225From2016.ix["2016",["High","Low"]].head(30) <DataFrame> ---')
print(n225.ix["2016",["High","Low"]].head(30))

# 2016年の　月初めの　高値と低値の値を表示
print('\n--- 2-1. n225From2016.resample(\'MS\').first() <DataFrame> ---')
print(n225.ix["2016",["High","Low"]].resample('MS').first())

# 2016年の　月最終日　高値と低値の値を表示
print('\n--- 2-2. n225From2016.resample(\'M\').last() <DataFrame> ---')
print(n225.ix["2016",["High","Low"]].resample('M').last())

# 2016年1月の　高値と低値の平均
print('\n--- 3-1. n225.ix["2016/1",["High","Low"]].mean() <Series> ---')
print(n225.ix["2016/1",["High","Low"]].mean())

# 2016年2月の　高値と低値の平均
print('\n--- 3-2. n225.ix["2016/2",["High","Low"]].mean() <Series> ---')
print(n225.ix["2016/2",["High","Low"]].mean())

# 2016年の月ごとの　高値と低値の平均　日付は月初めに寄せて表示
print('\n--- 3-3. n225.ix["2016",["High","Low"]].resample(\'MS\').mean().head(5) <DataFrame> ---')
print(n225.ix["2016",["High","Low"]].resample('MS').mean().head(5))

# 2016年の月ごとの　高値と低値の平均　日付は月終わりに寄せて表示
print('\n--- 3-4. n225.ix["2016",["High","Low"]].resample(\'M\').mean().head(5) <DataFrame> ---')
print(n225.ix["2016",["High","Low"]].resample('M').mean().head(5))

# 2016年の月ごとの　高値と低値の平均　日付は月終わりに寄せて表示
print('\n--- 3-5. n225.ix["2016",["High","Low"]].resample(\'M\').mean().tail(2) <DataFrame> ---')
print(n225.ix["2016",["High","Low"]].resample('M').mean().tail(2))
```
