```python
# NetworkX : ãƒãƒƒãƒˆãƒ¯ãƒ¼ã‚¯æ§‹é€ ã‚’åˆ†æžã™ã‚‹ãƒ©ã‚¤ãƒ–ãƒ©ãƒª
# ã‚°ãƒ©ãƒ•ç†è«–ã®ã‚¢ãƒ«ã‚´ãƒªã‚ºãƒ ã‚’æä¾›
%matplotlib notebook
import networkx as nx
import matplotlib.pyplot as pp

# ã‚°ãƒ©ãƒ•ã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆä½œæˆ
G = nx.Graph()

# ãƒŽãƒ¼ãƒ‰ã‚’è¿½åŠ 
#G.add_node(1) #1ã¤è¿½åŠ 
G.add_nodes_from(['A','B','C','D']) # ãƒªã‚¹ãƒˆã‹ã‚‰è¿½åŠ 
# edgeã‚’è¿½åŠ 
G.add_weighted_edges_from([('A', 'B',5),('A', 'C', 3),('B', 'C',6),('B', 'D',8),('C', 'D',9)])

# matplotlibã§æç”»
nx.draw(G, with_labels=True)
pp.show()

# ãƒ€ã‚¤ã‚¯ã‚¹ãƒˆãƒ©æ³• dijkstra_path æŽ¡ç‚¹çµŒè·¯ã‚’æ±‚ã‚ã‚‹ã‚¢ãƒ«ã‚´ãƒªã‚ºãƒ 
p1 = nx.dijkstra_path(G,'A','D')
print(p1)

# ãƒ€ã‚¤ã‚¯ã‚¹ãƒˆãƒ©æ³• dijkstra_path_length æŽ¡ç‚¹çµŒè·¯ã®æ‰€è¦æ™‚é–“
p2 = nx.dijkstra_path_length(G,'A','D')
print(p2)

# æœ€å°å…¨åŸŸæœ¨
mst = nx.minimum_spanning_tree(G)
print(mst.edges())
```
