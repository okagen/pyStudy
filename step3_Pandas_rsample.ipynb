```python
# resumpleメソッドの勉強
import pandas_datareader.data as pdr

start = "1949/5/6"
end = "2016/9/30"

# pandas_datareaderを使って、yahooからNIKKEI225のデータを読み込み 2016年以降にスライス <DataFrame>
price = pdr.DataReader("^N225", 'yahoo', start, end)
priceFrom2016 = price.ix["2016/1/1":]
print('\n--- 1. priceFrom2016 <DataFrame> ---')
print(priceFrom2016.resample('M').head(12))

# 2016年以降のNIKKEI225のデータ 1月の最初のレコード(2016/1/4)を取得 <Series>
print('\n--- 2-1. priceFrom2016.ix["2016/1/4"] <Sceries> ---')
print(priceFrom2016.ix["2016/1/4"])

# 2016年以降のNIKKEI225のデータ 1月の最初のレコード(2016/1/4)を取得 <Series>
print('\n--- 2-2. priceFrom2016.ix["2016/2/1"] <Sceries> ---')
print(priceFrom2016.ix["2016/2/1"])

# 2016年以降のNIKKEI225のデータ resampleを使って月ごとにし、各月の最初のレコードを取得 <DataFrame>
# データが無い場合でも、日付は各月の最初の日(１日)が表示される。
print('\n--- 2-3. priceFrom2016.resample(\'M\').first() <DataFrame> ---')
print(priceFrom2016.resample('M').first())


# 2016年以降のNIKKEI225のデータ resampleを使って月ごとにし、各月の最初のレコードを取得 <DataFrame>
# データが無い場合でも、日付は各月の最初の日(１日)が表示される。
print('\n--- 2-3. priceFrom2016.resample(\'M\').first().tail() <DataFrame> ---')
print(priceFrom2016.resample('M').first().tail())

# 2016年以降のNIKKEI225のデータ 1月の最後のレコード(2016/1/29)を取得 <Series>
print('\n--- 3-1. priceFrom2016.ix["2016/1/29"] <Sceries> ---')
print(priceFrom2016.ix["2016/1/29"])

# 2016年以降のNIKKEI225のデータ 1月の最後のレコード(2016/1/29)を取得 <Series>
print('\n--- 3-2. priceFrom2016.ix["2016/2/29"] <Sceries> ---')
print(priceFrom2016.ix["2016/2/29"])

# 2016年以降のNIKKEI225のデータ resampleを使って月ごとにし、各月の最後のレコードを取得 <DataFrame>
# 日付は各月の最終日(30日とか、31日とか)が表示される。
print('\n--- 3-3. priceFrom2016.resample(\'M\').last() <DataFrame> ---')
print(priceFrom2016.resample('M').last())
```
